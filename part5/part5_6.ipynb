```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid_function(x):
    """シグモイド関数

    Args:
        x (float): 引数

    Returns:
        float: 戻り値
    """
    return 1 / (1 + np.exp(-x))


def grad_sigmoid(x):
    """シグモイド関数の導関数

    Args:
        x (float): 引数

    Returns:
        float: 戻り値
    """
    y = sigmoid_function(x)
    return (1 - y) * y


x = np.linspace(-5, 5)
y = sigmoid_function(x)
y_grad = grad_sigmoid(x)

plt.plot(x, y, label="y")
plt.plot(x, y_grad, label="y_grad")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
import numpy as np


def approach_napier(n):
    """引数をネイピア数に近似します。

    Args:
        n (int): 引数

    Returns:
        float: ネイピア数の近似値
    """
    return (1 + 1 / n) ** n


n_list = [2, 4, 10, 100, 1000, 10000000]
for n in n_list:
    print("a_" + str(n) + " =", approach_napier(n))

print(f"e = {np.e}")
```

ネイピア数は以下の数式から求めることが出来る。
$$
a_n = \lim_{n \to ∞}(1 + \frac{1}{n})^n
