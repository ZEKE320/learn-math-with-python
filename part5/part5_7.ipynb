最急降下法
$$
x_i \leftarrow x_i - η\frac{∂f(\vec{x})}{∂x_i}
$$
ηは学習係数

```python
# 最急降下法の確認コード

import numpy as np
import matplotlib.pyplot as plt


def my_func(x):
    return x**2 - 2 * x


def grad_func(x):
    return 2 * x - 2


eta = 0.1
x = 4.0
record_x = list()
record_y = list()
for _ in range(20):
    y = my_func(x)
    record_x.append(x)
    record_y.append(y)
    x -= eta * grad_func(x)

x_f = np.linspace(-2, 4)
y_f = my_func(x_f)

plt.plot(x_f, y_f, linestyle="dashed")
plt.scatter(record_x, record_y)

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```

```python
# 最急降下法の失敗パターン(右の凹みは最小でない)
import numpy as np
import matplotlib.pyplot as plt


def my_func(x):
    return x**4 + 2 * x**3 - 3 * x**2 - 2 * x


def grad_func(x):
    return 4 * x**3 + 6 * x**2 - 6 * x - 2


def grad_decent(x):
    record_x = []
    record_y = []
    for _ in range(20):
        y = my_func(x)
        record_x.append(x)
        record_y.append(y)
        x -= eta * grad_func(x)
    plt.scatter(record_x, record_y)


eta = 0.01
x_arr = np.random.uniform(-2.8, 1.6, 5)

for x in x_arr:
    grad_decent(x)

x_f = np.linspace(-2.8, 1.6)
y_f = my_func(x_f)

plt.plot(x_f, y_f, linestyle="dashed")

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()


```
