```python
import numpy as np
import matplotlib.pyplot as plt


def random_int():
    return np.random.randint(10) + 1


arr_len = 100
arr = np.empty(arr_len, dtype=np.uint8).reshape(10, 10)
for i in range(arr_len):
    arr.flat[i] = random_int()

print(arr)

plt.hist(arr.flat, bins=10)
plt.show()
```
