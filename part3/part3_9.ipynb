```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi)
y_sin = np.abs(np.sin(x))
y_cos = np.abs(np.cos(x))

plt.scatter(x, y_sin, label="sin")
plt.scatter(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
