```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5)
y_1 = 2 * x
y_2 = 3 * x

# print(x)
# print(len(x))

plt.xlabel("x value", size=14)
plt.ylabel("y value", size=14)

plt.title("My Graph")

plt.grid()

plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2", linestyle="dashed")
plt.legend()
plt.show()
```
