```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    a = 3
    return x**a

x = np.linspace(0,2)
y = my_func(x)

plt.plot(x,y)
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def my_func(x):
    return np.sqrt(x)


x = np.linspace(0, 9)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def my_func(x):
    return np.sqrt(x) + 1


x = np.linspace(0, 4)
y = my_func(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
```
